# Reading and plotting temperature data

This notebook demonstrates how to plot the [HadCRUT4](https://crudata.uea.ac.uk/cru/data/temperature/) annual mean temperatures, using the [Julia](https://julialang.org) programming language. It also shows how easy it is to manipulate these graphs. You can adjust the plot parameters and execute cells using shift-enter (run all cells in order first).

## Load the data
Obtained from https://crudata.uea.ac.uk/cru/data/temperature/HadCRUT4-gl.dat

The mean temperature is at the end of each second line.

```julia
years = Int[]
T = Float64[]
for (i,l) in enumerate(eachline("HadCRUT4-gl.dat"))
    if i % 2 != 0
        spl = split(l)
        push!(years, parse(Int,spl[1]))
        push!(T, parse(Float64,spl[end]))
    end
end
```

## Directly plot the data

```julia
plot(years, T,
    legend=nothing,
    xlabel="Year", ylabel="Temperature anomaly (Â°C)",
    fmt=:png)
```

## Take only the last 20 years

```julia
plot(years[end-20:end], T[end-20:end],
    legend=nothing,
    xlabel="Year", ylabel="Temperature anomaly (Â°C)",
    fmt=:png)
```

## Add 14Â°C

```julia
plot(years[end-20:end], T[end-20:end].+14,
    legend=nothing,
    xlabel="Year",
    ylabel="Temperature (Â°C)", fmt=:png)
```

## Rescale the axes

```julia
plot(years[end-20:end], T[end-20:end].+14,
    legend=nothing,
    ylims=(12,16),
    xlabel="Year", ylabel="Temperature (Â°C)",
    fmt=:png)
```

## Convert to Kelvin

```julia
plot(years, T.+(14+273),
    legend=nothing,
    ylims=(0,300),
    xlabel="Year", ylabel="Temperature (K)",
    fmt=:png)
```
